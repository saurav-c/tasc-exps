```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def autolabel(rects, errors):
    for index, rect in enumerate(rects):
        height = rect.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, height - 4),
                    xytext=(0, 0),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color='white')
        error = height + errors[index]
        ax.annotate('{}'.format(round(error, 2)),
                    xy=(rect.get_x() + rect.get_width() / 2, error),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
error_kw = {'capsize': 5, 'capthick': 1, 'ecolor': 'black'}
```

```python
medians = np.array([[0.001995, 0.005106, 0.009185],[0.009013, 0.024895, 0.0466716]])
percentiles = np.array([[0.002331, 0.005667, 0.009942],[0.011795, 0.044433, 0.08975466]])
x_label = ["1 Write", "5 Writes", "10 Writes"]
```

```python
fig, ax = plt.subplots(figsize=(7,7))

width = 0.8
x = np.arange(len(x_label))
ax.set_xticks(x)
ax.set_xticklabels(x_label)

median_anna = medians[0] * 1000
percentile_anna = percentiles[0] * 1000
error_anna = [np.zeros(len(x_label)), percentile_anna - median_anna]

median_tasc = medians[1] * 1000
percentile_tasc = percentiles[1] * 1000
error_tasc = [np.zeros(len(x_label)), percentile_tasc - median_tasc]

plot_anna = ax.bar(x - width/8, median_anna, color='green', width=width/4, yerr=error_anna, error_kw=error_kw, align='center', label="Anna")
autolabel(plot_anna, error_anna[1])

plot_tasc = ax.bar(x + width/8, median_tasc, color='mediumaquamarine', width=width/4, yerr=error_tasc, error_kw=error_kw, align='center', label="TASC")
autolabel(plot_tasc, error_tasc[1])

ax.set_ylim(0, 100)
ax.legend()
ax.set_facecolor('lightblue')
ax.set_ylabel("Latency (ms)")

fig.show()
fig.savefig('writesvslatency.png')
```
